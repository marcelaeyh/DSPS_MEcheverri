```python
import pandas as pd
```

```python
earthquakes = pd.read_csv('https://raw.githubusercontent.com/fedhere/DSPS_FBianco/refs/heads/main/data/earthquakes.csv',sep=r'\s+')
earthquakes
```

```python
earthquakes.shape
```

```python
earthquakes.describe()
```

```python
earthquakes.dtypes
```

```python
earthquakes = earthquakes.rename({"#YYY/MM/DD":"date","HH:mm:SS.ss":"time","MAG":"mag"},axis=1)
```

```python
earthquakes
```

```python
pd.to_datetime("2002/12/29" + " " + "01:30:57.52")
```

```python
earthquakes["newdate"] = earthquakes["date"].map(pd.to_datetime)
```

```python
earthquakes["datetime"] = earthquakes["date"] + " " + earthquakes["time"]
earthquakes["datetime"]
```

```python
earthquakes["datetime"] = earthquakes["datetime"].map(pd.to_datetime)
earthquakes["datetime"]
```

```python
for i in earthquakes.index:
  if earthquakes.loc[i,'time'].endswith('60.00'):
    earthquakes.loc[i,'time'] = earthquakes.loc[i,'time'][:6] + "59.99"
```

```python
earthquakes["datetime"] = earthquakes["date"] + " " + earthquakes["time"]
earthquakes["datetime"] = earthquakes["datetime"].map(pd.to_datetime)
earthquakes["datetime"]
```
